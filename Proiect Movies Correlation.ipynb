```python
# Importam toate librariile de care avem nevoie pentru acest proiect
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Citim datele din fisierul csv
df = pd.read_csv(r'C:\Users\Alex\Desktop\movies.csv')
```

```python
#ne uitam la datele noastre
df
```

```python
# Verificam daca exista date lipsa
# Am creat o bucla for pentru a verifica daca exista ceva lipsa

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#stergem randurile cu valorile lipsa
df = df.dropna()
```

```python
#rulam iar pentru a verifica
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
```

```python
#Vedem ce tip de data are fiecare coloana
print(df.dtypes)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
#cream o noua coloana cu anul filmului pe baza coloanei released
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
#ordonam datele in functie de profitul brut pentru a vedea care filme au avut cele mai mari incasari
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#afisam toate valorile
pd.set_option('display.max_rows', None)
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#verificam daca exista duplicate
df['company'].sort_values(ascending=False)
#stergem duplicatele
df['company'] = df['company'].drop_duplicates()
#df['company'].sort_values(ascending=False)
```

```python
#Verificam corelatia dintre buget si profitul brut
#Am creat un scatter plot pentru bugetul si profitul brut

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Bugetul vs Profitul brut')
plt.xlabel('Bugetul pentru film')
plt.ylabel('Castigul brut')

plt.show()
```

```python
# creez o regresie pentru cele 2 variabile folosind seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#ne uitam la corelatiile dintre variabile
#alegelm unul dintre coeficientii de corelatie #pearson, #spearman, #kendall
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Matricea de corelatie pentru caracteristicie filmelor')
plt.xlabel('Caracteristicie filmelor')
plt.ylabel('Caracteristicie filmelor')

# avem o corelatie mare intre budget si gross -> 74.02% 
#respectiv intre votes si gross -> 61.47%
plt.show()
```

```python
df.head()
```

```python
#transformam variabilele nenumerice in unele categoriale pentru a putea gasi alte corelatii
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Matricea de corelatie pentru caracteristicie filmelor')
plt.xlabel('Caracteristicie filmelor')
plt.ylabel('Caracteristicie filmelor')

plt.show()
```

```python
df_numerized.corr()
```

```python
#folosim aceasta metoda pentru a putea extrage mai usor cele mai mari corelatii pentru fiecare caracteristica
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
